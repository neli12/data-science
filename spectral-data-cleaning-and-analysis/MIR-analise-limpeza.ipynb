```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
bd_mir = pd.read_csv('DB_MIR.csv', encoding ='ISO-8859-1', sep = ';')
bd_mir.shape
```

```python
bd_mir['State'] = bd_mir['State'].replace('Goais', 'Goias')
```

```python
bid_mir_A = bd_mir.loc[bd_mir['Layer'] == 'A']
bid_mir_A.shape
```

```python
cols = ['Type_Sample', 'Upper_depth', 'Lower_depth', 'Layer', 'Region_Number','Geology', 'pH.H2O', 'pH.KCl', 
        'Ca_mmol_kg', 'Mg_mmol_kg', 'K_mmol_kg', 'Na_mmol_kg', 'Al_mmol_kg', 'H_mmol_kg', 'H.Al_mmol_kg', 
        'CEC_pH7', 'CEC_ef', 'SB_mmol_kg', 'V_perc', 'm_perc', 'P_mg_kg', 'MO_gkg', 'Sand_gkg', 'Silt_gkg', 
        'Clay_gkg', 'Ha_Spec_cor', 'Hr_Spec_cor', 'Fe2O3_gkg', 'Classe.SiBCs.Confirmed', 'State.Abrev', 'Classe.SiBCs']
```

```python
bid_mir_A.drop(columns = cols, axis = 1, inplace = True)
```

```python
bid_mir_A.shape
```

```python
bid_mir_A.head()
```

```python
bid_mir_A.isnull().sum()
```
